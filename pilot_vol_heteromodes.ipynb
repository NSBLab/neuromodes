```python
from lapy.plot import plot_tet_mesh
from neuromodes.io import fetch_map
from neuromodes import EigenSolver
```

Initialise solver with surface and hetero

```python
neuromodes = 'C:/Users/ipop0003/neuromodes/neuromodes'

# Load "hetero map" (just slice a cortical map to fit hippocampal vertices)
myelinmap = fetch_map('myelinmap')[:1449]

volser = EigenSolver(f'{neuromodes}/data/hcp_hippo-lh_thr25.nii.gz.tetra.vtk', hetero=myelinmap, normalize=True).solve(100, seed=0)
print(volser)
```

```python
plot_tet_mesh(volser.geometry, vfunc=volser.emodes[:, 90], plot_edges=True)
```
